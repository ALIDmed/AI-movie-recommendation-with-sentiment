```python
import pandas as pd
import numpy as np
```

We are interested in 2017 films

```python
movies_metadata = pd.read_csv('../data/movies_metadata.csv')
movies_metadata.head()
```

```python
movies_metadata.columns
```

```python
movies_metadata = movies_metadata[['title', 'release_date', 'id', 'genres']]
movies_metadata.head()
```

```python
movies_metadata.release_date = pd.to_datetime(movies_metadata.release_date, errors='coerce')
```

```python
movies_metadata['year'] = movies_metadata.release_date.dt.year
```

```python
movies_metadata['year'].min(), movies_metadata['year'].max()
```

```python
movies_metadata = movies_metadata.loc[movies_metadata['year']==2017, ['title', 'id', 'genres', 'year']]
```

```python
movies_metadata['id'] = movies_metadata['id'].astype(int)
```

```python
movies_metadata.head()
```

```python
credits = pd.read_csv('../data/credits.csv')
```

```python
credits
```

```python
df = pd.merge(movies_metadata, credits, on='id')
```

```python
df
```

```python
df.loc[0, 'genres']
```

```python
import ast

df['genres'] = df['genres'].apply(ast.literal_eval)
df['cast'] = df['cast'].apply(ast.literal_eval)
df['crew'] = df['crew'].apply(ast.literal_eval)
```

```python
def extract_genres(x):
    genres = []
    for i in x:
        genres.append(i['name'])

    if genres == []:
        return np.NaN
    
    return " ".join(genres)
```

```python
df['genres'] = df['genres'].apply(extract_genres)
df['genres'] = df['genres'].str.replace('Science Fiction', 'Sci-Fi')
```

```python
def get_actor_by_num(x, num):
    if len(x) < num:
        return np.NaN
    return x[num-1].get('name', np.NaN)
```

```python
df['actor_1_name'] = df['cast'].apply(lambda x: get_actor_by_num(x, 1))
df['actor_2_name'] = df['cast'].apply(lambda x: get_actor_by_num(x, 2))
df['actor_3_name'] = df['cast'].apply(lambda x: get_actor_by_num(x, 3))
```

```python
def get_director_name(x):
    directors = []
    for i in x:
        if i['job'] == 'Director':
            directors.append(i['name'])

    if directors == []:
        return np.NaN

    return " ".join(directors)
```

```python
df['director_name'] = df['crew'].apply(get_director_name)
```

```python
df = df[['title', 'genres', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'director_name']]
```

```python
df.rename(columns={'title':'movie_title'}, inplace=True)
df.movie_title = df.movie_title.str.lower()
```

```python
old = pd.read_csv('../data/data.csv')
```

```python
new = pd.concat([old, df], ignore_index=True)
```

```python
new
```

```python
new.to_csv('../data/new_data.csv', index=False)
```
